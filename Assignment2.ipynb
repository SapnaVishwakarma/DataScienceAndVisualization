```python
# Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_target = pd.Series(iris.target, name='target')

# Split the dataset into training and testing sets with an 80-20 split
X_train, X_test, y_train, y_test = train_test_split(iris_df, iris_target, test_size=0.2, random_state=42)

# Print the number of samples in both the training and testing sets
num_train_samples = X_train.shape[0]
num_test_samples = X_test.shape[0]
print(f"Number of samples in the training set: {num_train_samples}")
print(f"Number of samples in the testing set: {num_test_samples}")

# Visualize the data split using bar charts
labels = ['Training Set', 'Testing Set']
sample_counts = [num_train_samples, num_test_samples]

plt.figure(figsize=(4, 3))
plt.bar(labels, sample_counts, color=['blue', 'green'])
plt.xlabel('Dataset')
plt.ylabel('Number of Samples')
plt.title('Number of Samples in Training and Testing Sets')
plt.show()
```
